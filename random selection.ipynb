Random selection for 1.1

```python
import os
import shutil
import random

def sample_images(source_folder, destination_folder, x):

    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    files = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]

    sampled_files = random.sample(files, x)

    for file in sampled_files:
        shutil.move(os.path.join(source_folder, file), os.path.join(destination_folder, file))

source_folder = '/content/gdrive/MyDrive/ACV/yolov9/runs/detect/exp/crops/person'
destination_folder = '/content/1.1'
x = 12  # Number of images to sample

sample_images(source_folder, destination_folder, x)

source_folder = '/content/gdrive/MyDrive/ACV/yolov9/runs/detect/exp2/crops/person'
destination_folder = '/content/1.1'
x = 13  # Number of images to sample

sample_images(source_folder, destination_folder, x)

source_folder = '/content/gdrive/MyDrive/ACV/yolov9/runs/detect/exp3/crops/person'
destination_folder = '/content/1.1'
x = 12  # Number of images to sample

sample_images(source_folder, destination_folder, x)

source_folder = '/content/gdrive/MyDrive/ACV/yolov9/runs/detect/exp4/crops/person'
destination_folder = '/content/1.1'
x = 13  # Number of images to sample

sample_images(source_folder, destination_folder, x)
```

```python
import os
import shutil
import random

def copy_matched_files(base_folders, pose_estimation_subfolder, target_folders, destination_root):
    # Ensure destination folders exist
    for folder in target_folders:
        dest_path = os.path.join(destination_root, folder.strip('/'))
        if not os.path.exists(dest_path):
            os.makedirs(dest_path)

    # Iterate over each base folder
    for base_folder in base_folders:
        # Source folder for pose estimation images
        source_folder = os.path.join(base_folder, pose_estimation_subfolder)
        print(f"Checking source folder: {source_folder}")  # Debug

        # Check if the source folder exists
        if not os.path.exists(source_folder):
            print(f"Source folder not found: {source_folder}")
            continue

        # Iterate over each target folder
        for target_folder in target_folders:
            target_path = os.path.join(base_folder, target_folder.strip('/'))
            print(f"Checking target folder: {target_path}")  # Debug

            if not os.path.exists(target_path):
                print(f"Target folder not found: {target_path}")
                continue

            # List all files in the target subdirectory
            files = [f for f in os.listdir(target_path) if os.path.isfile(os.path.join(target_path, f))]
            print(f"Found files in {target_path}: {files}")  # Debug

            # Copy files with matching names from the pose estimation folder
            for file in files:
                source_file_path = os.path.join(source_folder, file)
                if os.path.exists(source_file_path):
                    destination_path = os.path.join(destination_root, target_folder.strip('/'), file)
                    shutil.copy(source_file_path, destination_path)
                    print(f"Copied {source_file_path} to {destination_path}")  # Debug
                else:
                    print(f"No matching file found for {file} in {source_folder}")

# Example usage
base_folders = [
    '/content/gdrive/MyDrive/ACV/openpose/TheSopranos',
    '/content/gdrive/MyDrive/ACV/openpose/TheIrishman',
    '/content/gdrive/MyDrive/ACV/openpose/TheGodfather',
    '/content/gdrive/MyDrive/ACV/openpose/MafiaVideogame'
]
pose_estimation_subfolder = 'pose-estimation'
target_folders = [
    '/train/full-body_front_view',
    '/train/full-body_back_view',
    '/train/head-and-shoulder_front_view',
    '/train/head-and-shoulder_back_view',
    '/train/others'
]
destination_root = '/content'

copy_matched_files(base_folders, pose_estimation_subfolder, target_folders, destination_root)
```

```python
import cv2
import os

def extract_frames(video_path, output_folder):

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)


    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error opening video file")
        return


    fps = int(cap.get(cv2.CAP_PROP_FPS))
    print(f"Frames per second: {fps}")

    frame_count = 0
    saved_frame_count = 0

    while True:
        ret, frame = cap.read()

        if not ret:
            print("Can't receive frame (stream end?). Exiting ...")
            break


        if frame_count % fps == 0:
            frame_filename = os.path.join(output_folder, f"frame_{saved_frame_count}.jpg")
            cv2.imwrite(frame_filename, frame)
            saved_frame_count += 1
            print(f"Saved {frame_filename}")

        frame_count += 1

    cap.release()
    print(f"Total frames saved: {saved_frame_count}")

extract_frames('/content/gdrive/MyDrive/ACV/Data/Train/game/MafiaVideogame.mp4', '/content/Training/games/class1')
extract_frames('/content/gdrive/MyDrive/ACV/Data/Train/movie/TheGodfather.mp4', '/content/Training/movies/class2')
extract_frames('/content/gdrive/MyDrive/ACV/Data/Train/movie/TheIrishman.mp44', '/content/Training/movies/class2')
extract_frames('/content/gdrive/MyDrive/ACV/Data/Train/movie/TheSopranos.mp4', '/content/Training/movies/class2')
```
